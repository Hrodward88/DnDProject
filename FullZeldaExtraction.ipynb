```python
def releaselimpia (input_str):

    year_pattern = r'\b\d{4}\b'
    system_pattern = r'\s(.*?)(?=\s\d{4}\s|\s*$)'

    years = re.findall(year_pattern, input_str)
    systems = re.findall(system_pattern, input_str)

    output_list = [f'{year} - {system}' for year, system in zip(years, systems)]

    return output_list
```

```python
def get_score(string, pattern=r'^\d+\. (.+) \(Score- (\d+)\)$'):
    """
    Extracts the name and score from a string using the provided regular expression pattern.
    
    Args:
        string (str): The input string to extract the name and score from.
        pattern (str): The regular expression pattern to use for matching the string.
        
    Returns:
        A tuple containing the name and score extracted from the input string.
    """
    match = re.search(pattern, string)
    if match:
        name = match.group(1)
        score = int(match.group(2))
        return (name, score)
    else:
        return None
```
